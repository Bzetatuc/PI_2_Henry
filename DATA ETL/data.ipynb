```python
 #import modules
import pandas as pd
import datetime as dt
import time as t
from pycoingecko import CoinGeckoAPI

#create a client
cg = CoinGeckoAPI()

#confirm connection
cg.ping()

#get a list of coins, sort df by id
coinList = cg.get_coins_list()
coinDataFrame = pd.DataFrame.from_dict(coinList).sort_values('id'
                                      ).reset_index(drop=True)

coins = ['binancecoin', 'bitcoin', 'cardano', 'dogecoin', 'ethereum', 'polkadot', 'ripple', 'solana', 'tether', 'usd-coin']

#get list of supported VS currencies
counterCurrencies = cg.get_supported_vs_currencies()
vsCurrencies = ['usd']

#most simple price request - nested dictionary format
complexPriceRequest = cg.get_price(ids=coins, 
                        vs_currencies=vsCurrencies, 
                        include_market_cap=True,
                        include_24hr_vol=True,
                        include_24hr_change=True,
                        include_last_updated_at=True)

# Create DataFrames for the prices
complex_prices_df = pd.DataFrame(complexPriceRequest)

# Display the DataFrames for the specified coins

print("\nComplex Prices:")
print(complex_prices_df.to_string())

# Get data on coin categories
coinCategoriesData = pd.DataFrame(cg.get_coins_categories(order='market_cap_desc'))

# Dictionary to DataFrame
coinCategoriesDataFrame = pd.DataFrame.from_dict(coinCategoriesData).sort_values('id').reset_index(drop=True)

# By category, see cg.get_coins_categories_list(), in order
coinsMktDataByCategory = cg.get_coins_markets(vs_currency='usd',
                                               category='stablecoins',
                                               order='volume_desc')

# Create a DataFrame for coins market data by category
coinsMktDataByCategory_df = pd.DataFrame(coinsMktDataByCategory)

# Display the DataFrame for coins market data by category
print("\nCoins Market Data by Category:")
print(coinsMktDataByCategory_df.to_string())
```

```python
def get_historical_data(cryptocurrency, fiat_currency, number_of_days):
    historic_price = coingecko.get_coin_market_chart_by_id(cryptocurrency, fiat_currency, number_of_days)
    prices = [price[1] for price in historic_price['prices']]
    return prices
```

```python
from datetime import datetime, date

def calcular_dias_desde_inicio():
    fecha_inicio = date(2019, 12, 31)
    fecha_actual = date.today()
    dias_transcurridos = (fecha_actual - fecha_inicio).days
    return dias_transcurridos

# Llamar a la funcin para obtener el resultado
dias_desde_inicio = calcular_dias_desde_inicio()
print("Das desde el 31 de diciembre de 2019 hasta hoy:", dias_desde_inicio)
```

```python

# Obtener datos histricos de precios y capitalizacin de mercado para Bitcoin (BTC)
bitcoin = cg.get_coin_market_chart_range_by_id(id='bitcoin', vs_currency='usd', from_timestamp=1577761200, to_timestamp=1692367855)
historical_prices = bitcoin['prices']
market_caps = bitcoin['market_caps']

# Obtener informacin bsica de la moneda bitcoin
token_info = cg.get_coin_by_id(id='bitcoin')
symbol = token_info['symbol']
name = token_info['name']

# Crear un DataFrame con los datos histricos de precios
historical_df = pd.DataFrame(historical_prices, columns=['timestamp', 'price'])
historical_df['timestamp'] = pd.to_datetime(historical_df['timestamp'], unit='ms').dt.date  # Convertir a fecha

# Crear un DataFrame con los datos histricos de capitalizacin de mercado
market_cap_df = pd.DataFrame(market_caps, columns=['timestamp', 'market_cap'])
market_cap_df['timestamp'] = pd.to_datetime(market_cap_df['timestamp'], unit='ms').dt.date  # Convertir a fecha

# Combinar los DataFrames utilizando join
combined_df_bitcoin = historical_df.join(market_cap_df.set_index('timestamp'), on='timestamp')

# Calcular el valor ATH y su fecha correspondiente
ath_value = combined_df_bitcoin['price'].max()
ath_date = combined_df_bitcoin.loc[combined_df_bitcoin['price'].idxmax()]['timestamp']

# Agregar la columna ATH al DataFrame
combined_df_bitcoin['ath_value'] = ath_value
combined_df_bitcoin['ath_date'] = ath_date

# Agregar las columnas de symbol y name directamente al DataFrame
combined_df_bitcoin['symbol'] = symbol
combined_df_bitcoin['name'] = name

# Mostrar el DataFrame con las nuevas columnas
combined_df_bitcoin

```

```python
max_price = combined_df_bitcoin['price'].max()

# Mostrar el valor mximo de 'price' en la salida
print("Monto mximo de la columna 'price':", max_price)
```

```python
# Cambiar el nombre de la columna 'timestamp' a 'date'
combined_df_bitcoin = combined_df_bitcoin.rename(columns={'timestamp': 'date'})
```

```python
cardano = cg.get_coin_market_chart_range_by_id(id='cardano', vs_currency='usd', from_timestamp=1577761200, to_timestamp=1692367855)
historical_prices = cardano['prices']
market_caps = cardano['market_caps']

# Obtener informacin bsica de la moneda cardano
token_info = cg.get_coin_by_id(id='cardano')
symbol = token_info['symbol']
name = token_info['name']

# Crear un DataFrame con los datos histricos de precios
historical_df = pd.DataFrame(historical_prices, columns=['timestamp', 'price'])
historical_df['timestamp'] = pd.to_datetime(historical_df['timestamp'], unit='ms').dt.date  # Convertir a fecha

# Crear un DataFrame con los datos histricos de capitalizacin de mercado
market_cap_df = pd.DataFrame(market_caps, columns=['timestamp', 'market_cap'])
market_cap_df['timestamp'] = pd.to_datetime(market_cap_df['timestamp'], unit='ms').dt.date  # Convertir a fecha

# Combinar los DataFrames utilizando join
combined_df_cardano = historical_df.join(market_cap_df.set_index('timestamp'), on='timestamp')

# Calcular el valor ATH y su fecha correspondiente
ath_value = combined_df_cardano['price'].max()
ath_date = combined_df_cardano.loc[combined_df_cardano['price'].idxmax()]['timestamp']

# Agregar la columna ATH al DataFrame
combined_df_cardano['ath_value'] = ath_value
combined_df_cardano['ath_date'] = ath_date

# Agregar las columnas de symbol y name directamente al DataFrame
combined_df_cardano['symbol'] = symbol
combined_df_cardano['name'] = name

# Mostrar el DataFrame con las nuevas columnas
combined_df_cardano
```

```python
# Cambiar el nombre de la columna 'timestamp' a 'date'
combined_df_cardano = combined_df_cardano.rename(columns={'timestamp': 'date'})

# Guardar el DataFrame en un archivo CSV con el nuevo nombre de columna
combined_df_cardano.to_csv('combined_df_cardano.csv', index=False)
```
