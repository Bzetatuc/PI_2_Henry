```python
 #import modules
import pandas as pd
import datetime as dt
import time as t
from pycoingecko import CoinGeckoAPI

#create a client
cg = CoinGeckoAPI()

#confirm connection
cg.ping()

#get a list of coins, sort df by id
coinList = cg.get_coins_list()
coinDataFrame = pd.DataFrame.from_dict(coinList).sort_values('id'
                                      ).reset_index(drop=True)

coins = ['binancecoin', 'bitcoin', 'cardano', 'dogecoin', 'ethereum', 'polkadot', 'ripple', 'solana', 'tether', 'usd-coin']

#get list of supported VS currencies
counterCurrencies = cg.get_supported_vs_currencies()
vsCurrencies = ['usd']

#most simple price request - nested dictionary format
complexPriceRequest = cg.get_price(ids=coins, 
                        vs_currencies=vsCurrencies, 
                        include_market_cap=True,
                        include_24hr_vol=True,
                        include_24hr_change=True,
                        include_last_updated_at=True)

# Create DataFrames for the prices
complex_prices_df = pd.DataFrame(complexPriceRequest)

# Display the DataFrames for the specified coins

print("\nComplex Prices:")
print(complex_prices_df.to_string())

# Get data on coin categories
coinCategoriesData = pd.DataFrame(cg.get_coins_categories(order='market_cap_desc'))

# Dictionary to DataFrame
coinCategoriesDataFrame = pd.DataFrame.from_dict(coinCategoriesData).sort_values('id').reset_index(drop=True)

# By category, see cg.get_coins_categories_list(), in order
coinsMktDataByCategory = cg.get_coins_markets(vs_currency='usd',
                                               category='stablecoins',
                                               order='volume_desc')

# Create a DataFrame for coins market data by category
coinsMktDataByCategory_df = pd.DataFrame(coinsMktDataByCategory)

# Display the DataFrame for coins market data by category
print("\nCoins Market Data by Category:")
print(coinsMktDataByCategory_df.to_string())
```

```python
coins = ['bitcoin', 'ethereum','tether', 'binancecoin', 'ripple', 'usd-coin', 'dogecoin', 'cardano', 'solana', 'polkadot']

# By category, see cg.get_coins_categories_list(), in order
coinsMktDataByCategory = cg.get_coins_markets(vs_currency='usd',
                                               ids=coins,
                                               order='volume_desc')

# Create a DataFrame for coins market data by category
coinsMktDataByCategory_df = pd.DataFrame(coinsMktDataByCategory)

# Display the DataFrame for coins market data by category
print("\nCoins Market Data by Category:")
print(coinsMktDataByCategory_df.to_string())
```

```python
# Ordenar el DataFrame por la columna "market_cap" de mayor a menor
coinsMktDataByCategory_df = coinsMktDataByCategory_df.sort_values(by='market_cap', ascending=False)

# Mostrar el DataFrame de precios complejos
print("\nComplex Prices (Sorted by Market Cap):")
print(coinsMktDataByCategory_df.to_string())
```

```python
coinsMktDataByCategory_df = coinsMktDataByCategory_df[['id', 'symbol', 'name', 'current_price', 'market_cap','market_cap_rank','high_24h', 'low_24h', 'price_change_24h',
       'price_change_percentage_24h', 'market_cap_change_24h','market_cap_change_percentage_24h', 'ath', 'ath_change_percentage',
       'ath_date']]
```

```python
coinsMktDataByCategory_df
```

```python
formatted_cols = ['current_price', 'market_cap', 'high_24h', 'low_24h', 'price_change_24h', 
                  'market_cap_change_24h', 'ath', 'ath_change_percentage']
```

```python
coinsMktDataByCategory_df[formatted_cols] = coinsMktDataByCategory_df[formatted_cols].applymap(lambda x: '{:,.2f}'.format(x) if isinstance(x, (int, float)) else x)
```

```python
percentage_cols = ['price_change_percentage_24h', 'market_cap_change_percentage_24h', 'ath_change_percentage']
```

```python
# Convertir las columnas de porcentaje a valores numricos (suponiendo que los porcentajes estn en formato 'xx.xx%')
coinsMktDataByCategory_df[percentage_cols] = coinsMktDataByCategory_df[percentage_cols].applymap(lambda x: float(x.strip('%')) / 100 if isinstance(x, str) else x)

# Aplicar el formato de porcentaje a las columnas convertidas
coinsMktDataByCategory_df[percentage_cols] = coinsMktDataByCategory_df[percentage_cols].applymap('{:.2%}'.format)
```
